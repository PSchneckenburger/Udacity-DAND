## Introduction


> For this project, I decided to use the provided Dataset **No-show appointments.csv**
> This dataset collected informations from 100k medical appointments in Brazil and is focused on  the question of whether or not patients show up for their appointment.
>
> The original source: https://www.kaggle.com/joniarroba/noshowappointments
>
>The question the author has is, if it is predictable if a patient doesn't attend to his appointment.
>
> A number of characteristics about the appointments and the patients are included in each row:
>*  **PatientId** - Identification of a patient/n
>* **AppointmentID** - Identification of each appointment
>* **Gender** - Male or Female
>* **ScheduledDay** - on what day the patients set up their appointment.
>* **AppointmentDay** - The day of the actuall appointment, when they have to visit the doctor. 
>* **Age** - How old is the patient. 
>* **Neighbourhood** - Where the appointment takes place. 
>* **Scholarship** - whether the patient gets financial support, more details: https://en.wikipedia.org/wiki/Bolsa_Fam%C3%ADlia 
>* **Hipertension** - if patient suffers from hypertension
>* **Diabetes** - if patient suffers from diabetes
>* **Alcoholism**  -if patient suffers from alcoholism
>* **Handcap** - number of handicaps a patient has
>* **SMS_received** - if 1 or more messages was sent to the patient 
>* **No-show** - if patient showed up, whereas 'no' means he showed up, 'yes' means he didn't show!

> Intention for this analysis:
>* Can factors be identified that will be useful as predictors for missing medical appointments?
>* Are there means to improve the compliance?

### Converting data types and removing errant data

```python
data = df

# changing column names
data.rename(columns = {'PatientId':'PatientID', 'Hipertension':'Hypertension', 'Handcap':'Handicap', 'SMS_received':'SMS', 'No-show':'No_show'}, inplace=True )
```

```python
# changing scientific notation in PatientID to integer, just for nicer view
data['PatientID'] = data['PatientID'].astype(np.int64)
# using astype(int) results in negative values
# https://stackoverflow.com/questions/56391169/pandas-astypeint-applied-to-float-column-returns-negative-numbers

# change datatypes of SceduledDay and AppointmentDay to datetime
data['ScheduledDay'] = data['ScheduledDay'].apply(np.datetime64)
data['AppointmentDay'] = data['AppointmentDay'].apply(np.datetime64)

# extracting difference between ScheduledDay and AppointmentDay in number of days and creating new column
data['Delay'] = data.AppointmentDay - data.ScheduledDay.dt.normalize()
data['Delay'] = data['Delay'].dt.days

#change values for No_show to 0 (Patient showed up) and 1 (Patient didn't show up)
data['No_show']= data['No_show'].apply(lambda x: 0 if x =="No" else 1)

# change values in Handicap if there is at least one or none to 1 and 0 respectively
data['Handicap']=data['Handicap'].apply(lambda x: 1 if x >1 else x)


data.head()
```

```python
# removing row with negative Age
data.drop(negative_age.index, inplace = True)
#data.drop(data.index[99832], inplace=True)

# control
negative_age=data.query('Age < 0')
negative_age
```

```python
# checking for errors in dates
negativedays = data[data['Delay'] < 0]
negativedays
```

There are in fact few rows where ScheduledDay was later than AppointmentDay, they have to be omitted

```python
# stripping rows with date errors
data.drop(negativedays.index, inplace=True)
#control
negativedays = data[data['Delay'] < 0]
negativedays
```

```python
# looking for possibly doubled patient data
# https://stackoverflow.com/questions/46640945/grouping-by-multiple-columns-to-find-duplicate-rows-pandas   
doubled_data = data[data.duplicated(subset=['PatientID','ScheduledDay', 'AppointmentDay'], keep=False)]
doubled_data.head(10)
```

> Not all the the values were identical, there sometimes were differences in the variables *SMS* and *No_show*

```python
# check for variables with different values in the doubled rows
check = doubled_data.groupby('PatientID')['SMS', 'No_show'].mean()
check.head(10)
```

>Obviously, there were entries for the same appointment with different values vor SMS sent and No_show respectively, but not for all patients.  The reason for different entries could be that the patients dind't show off, were sent an SMS and then come to the appointments or not, but there also were patients without sent an SMS, so no reason to have doubled entries.
>Thus I decided to eliminate rows that were doubled in all variables.

```python
# drop AppointmentID, since it is not needed
data.drop('AppointmentID', axis=1, inplace=True)
# number of duplicated rows
print(data.duplicated().sum())
# delete duplicated rows
data.drop_duplicates(inplace=True)
# control
data.duplicated().sum()
```

### Is sending an SMS a good mean to prevent patients from missing  appointments? ###


```python
# counts by SMS and No_show
data.groupby('SMS')['No_show'].value_counts()
```

> There were much more cases, where no SMS was sent than sent.
> For better comparability it would be better to regard the propotions of kept or not kept 
> appointments to the total of not sent or sent SMS 

```python
# get the total count of appointments, where SMS were sent and not sent
totals = data.groupby('SMS').count()['AppointmentDay']
# get the counts of appointments according to SMS sent/not sent and showed/not showed
counts = data.groupby(['SMS', 'No_show']).count()['AppointmentDay']
# calculation proportions
SMS_counts = counts[1]
NoSMS_counts = counts[0]
SMS_prop = SMS_counts/totals[1]
NoSMS_prop = NoSMS_counts/totals[0]

#creating the plot
ind = np.arange(len(NoSMS_prop))  # the x locations for the groups
width = 0.35       # the width of the bars
# plot bars
NoSMS_bars = plt.bar(ind, NoSMS_prop, width, color='g', alpha=.7, label='No SMS');
SMS_bars = plt.bar(ind + width, SMS_prop, width, color='r', alpha=.9, label='SMS');

# title and labels
plt.ylabel('proportion of Appointments')
plt.xlabel('Appointment')
plt.title('Effect of SMS on keeping appointment')
locations = ind + width / 2  # xtick locations
labels = ['kept ', 'missed' ]  # xtick labels
plt.xticks(locations, labels)

# legend
plt.legend();
```

>Regarding the plot, it seems that Patients receiving an SMS were more likely to not keep their appointments. This is unexpected and should be examined more.

```python
# comparing the groups that received sms with the group that didn't receive sms
data.groupby('SMS').mean()
```

> The main difference between the two groups is the delay: the mean delay in the SMS group is 19,0 whereas in the other group is 5.9.

### Does the waitint time between scheduling and the appointment itself have an influence on how likely an appointment is missed? ###


>For this and the following questions, only the data where no SMS was sent, will be used, to eliminate reminder interference



```python
# creating datasubset where no SMS was sent
noSMS_data = data.query('SMS==0')

noSMS_data.describe()
```

```python
noSMS_data.Delay.plot(kind='hist', figsize =(10,5));
plt.title('Distribution of Delay in noSMS group');
plt.xlabel('Delay')
plt.ylabel('Counts');
```

```python
# distribution of delay regarding to keeping/missing the appointment
sns.set(rc={'figure.figsize':(10,5)})

g = sns.boxplot(x='No_show', y='Delay', data=noSMS_data);
plt.title('Delay in regard of keeping appointments in noSMS group');
plt.xlabel('Appointments');
plt.ylabel('Delay');
g.set(xticklabels=['Kept', 'missed']);
```

> As clearly can be seen, appointments with smaller delays are more likely to be kept.

### Which influence does patients age have on missing appointments?###

The distribution of Age is similar in the noSMS group than in the overall group

### Does gender have an influence on keeping appointments?###

```python
# distribution of gender in noSMS group
noSMS_data.Gender.value_counts()
```

```python
noSMS_data.Gender.value_counts().plot(kind='pie', figsize=(4,4));
```

The distribution of gender in the noSMS group obviously is similar to that of the total group.

```python
# get the total count of appointments by gender
gender_totals = noSMS_data.groupby('Gender').count()['AppointmentDay']
# get the counts of appointments according to gender and showed/not showed
gender_counts = noSMS_data.groupby(['Gender', 'No_show']).count()['AppointmentDay']

# calculation proportions
females = gender_counts['F']
males = gender_counts['M']
females_prop = females/gender_totals['F']
males_prop = males/gender_totals['M']

#creating the plot
ind = np.arange(len(females_prop))  # the x locations for the groups
width = 0.35       # the width of the bars
# plot bars
males_bars = plt.bar(ind, males_prop, width, color='g', alpha=.7, label='males');
females_bars = plt.bar(ind + width, females_prop, width, color='r', alpha=.9, label='females');

# title and labels
plt.ylabel('proportion of Appointments')
plt.xlabel('Appointments')
plt.title('Effect of gender on keeping appointments in the noSMS group')
locations = ind + width / 2  # xtick locations
labels = ['kept ', 'missed' ]  # xtick labels
plt.xticks(locations, labels)

# legend
plt.legend();
```

> There is only a minimal difference between males and females on keeping appointments, probably not being significant. 

> ### Are some medical issues more likely to cause missing appointments than others? ###

### Do attending in a social program might improve the compliance?###

```python
# distribution of scholarship
noSMS_data.Scholarship.value_counts().plot(kind='pie', figsize=(3,3));
```

```python
noSMS_data.Scholarship.value_counts()
```

>Only 7797 Appointments were within a social program, that is less than 10 percent.
For clearer comparison, here also the data were counted as proportions to the group size.

```python
# get the total count of appointments, where Patient has a Scholarship
scholar_totals = noSMS_data.groupby('Scholarship').count()['AppointmentDay']
# get the counts of appointments according to scholarship and showed/not showed
scholar_counts = noSMS_data.groupby(['Scholarship', 'No_show']).count()['AppointmentDay']

# calculation proportions
counts = scholar_counts[1]
no_counts = scholar_counts[0]
scholar_prop = counts/scholar_totals[1]
noscholar_prop = no_counts/scholar_totals[0]

#creating the plot
ind = np.arange(len(noscholar_prop))  # the x locations for the groups
width = 0.35       # the width of the bars
# plot bars
noscholar_bars = plt.bar(ind, noscholar_prop, width, color='g', alpha=.7, label='No scholarship');
scholar_bars = plt.bar(ind + width, scholar_prop, width, color='r', alpha=.9, label='scholarship');

# title and labels
plt.ylabel('Proportion')
plt.xlabel('Appointments')
plt.title('Effect of scholarship on keeping appointment in the noSMS group')
locations = ind + width / 2  # xtick locations
labels = ['kept ', 'missed' ]  # xtick labels
plt.xticks(locations, labels)

# legend
plt.legend();
```

> Patients that have a scholarship in the social program seem to keep their appointments slightly less than others.

## Conclusions

>* Overall, only 20% of the appointments were missed by the patients.

> Influence of several factors on missing appointments:
>* One main factor for missing appointments is the delay between scheduling and the appointment
>* Appointments are more likely to be missed by adolescents and young adults than babys (where caring adults shurely make appointments) ore elder persons
>* Among the medical issues reported, people with hypertension seem to miss their appointments more often than people with other issues.
>* Gender has no obvious effect on keeping or missing appointments
>* Participants of the Bolsa program seem to be slightly more likely to miss their appointments. 

> Regarding effect of SMS to improve compliance:
>* Although on first glimpse sending SMS would not improve the rate of keeping appointments, this may be due to the fact, that SMS were sent more in case of longer delays. Therefore sending SMS might nevertheless be a good way to improve compliance. 


> Limitations:
>* To prove the effect of sending SMS as reminder, it would be necessary to compare the SMS group with a delay controlled set out of the noSMS group.
>* As reported in the data wrangling section there were possibly errant data in form of doubled entries. The validity of these entries should be cleared.
